```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
df = pd.read_csv('..\Data\Walking.csv')

# Create a figure and a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Define the y-axis limits
y_min = 14  # Lower bound for y-axis
y_max = 16.5  # Upper bound for y-axis

# Define the x-axis zoom limits
x_min = 700  # Lower bound for x-axis
x_max = 900  # Upper bound for x-axis

# Plot s1 in the first subplot using the index as the x-axis
axs[0, 0].plot(df.index, df['s1'], color='r')
axs[0, 0].set_title('Meta 1-2')
axs[0, 0].set_xlabel('Data Point Index')
axs[0, 0].set_ylabel('Pressure Sensor s1')
axs[0, 0].set_ylim(y_min, y_max)  # Set y-axis limits
axs[0, 0].set_xlim(x_min, x_max)  # Set x-axis limits (zooming)
axs[0, 0].grid(True)

# Plot s2 in the second subplot using the index as the x-axis
axs[0, 1].plot(df.index, df['s2'], color='g')
axs[0, 1].set_title('Toe')
axs[0, 1].set_xlabel('Data Point Index')
axs[0, 1].set_ylabel('Pressure Sensor s2')
axs[0, 1].set_ylim(y_min, y_max)  # Set y-axis limits
axs[0, 1].set_xlim(x_min, x_max)  # Set x-axis limits (zooming)
axs[0, 1].grid(True)

# Plot s3 in the third subplot using the index as the x-axis
axs[1, 0].plot(df.index, df['s3'], color='b')
axs[1, 0].set_title('Meta 3-4')
axs[1, 0].set_xlabel('Data Point Index')
axs[1, 0].set_ylabel('Pressure Sensor s3')
axs[1, 0].set_ylim(y_min, y_max)  # Set y-axis limits
axs[1, 0].set_xlim(x_min, x_max)  # Set x-axis limits (zooming)
axs[1, 0].grid(True)

# Plot s4 in the fourth subplot using the index as the x-axis
axs[1, 1].plot(df.index, df['s4'], color='m')
axs[1, 1].set_title('Heel')
axs[1, 1].set_xlabel('Data Point Index')
axs[1, 1].set_ylabel('Pressure Sensor s4')
axs[1, 1].set_ylim(y_min, y_max)  # Set y-axis limits
axs[1, 1].set_xlim(x_min, x_max)  # Set x-axis limits (zooming)
axs[1, 1].grid(True)

# Adjust layout to prevent overlapping of labels and titles
plt.tight_layout()

# Now create a new figure for the combined plot of all sensors
plt.figure(figsize=(10, 6))

# Plot all sensors on the same axis
plt.plot(df.index, df['s1'], label='Meta 1-2 (s1)', color='r')
plt.plot(df.index, df['s2'], label='Toe (s2)', color='g')
plt.plot(df.index, df['s3'], label='Meta 3-4 (s3)', color='b')
plt.plot(df.index, df['s4'], label='Heel (s4)', color='k')

# Set titles and labels
plt.title('Combined Pressure Sensors')
plt.xlabel('Data Point Index')
plt.ylabel('Pressure Sensor Readings')

# Set x and y limits
plt.ylim(y_min, y_max)
plt.xlim(x_min, x_max)

# Add a legend to differentiate between the plots
plt.legend()

# Add grid for better visualization
plt.grid(True)

# Show both plots
plt.show()
```
